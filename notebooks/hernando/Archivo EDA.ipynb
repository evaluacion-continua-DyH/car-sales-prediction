```python
import pandas as pd

df = pd.read_csv("../../data/car_sales_dataset.csv")
df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2. InformaciÃ³n general
# ================================

print("ðŸŸ¦ SHAPE (filas, columnas):")
print(df.shape)

print("\nðŸŸ¦ Tipos de datos:")
print(df.dtypes)

print("\nðŸŸ¦ Valores nulos por columna:")
print(df.isnull().sum())

print("\nðŸŸ¦ InformaciÃ³n completa:")
df.info()
```

```python
# 3. EstadÃ­sticas descriptivas
# ================================

df.describe().T
```

```python
# 5. DistribuciÃ³n de variables numÃ©ricas
# ================================

df.hist(figsize=(12,8), bins=30)
plt.tight_layout()
plt.show()
```

```python
# 7. Matriz de correlaciÃ³n numÃ©rica
# ================================

df_numeric = df.select_dtypes(include=['int64','float64'])
df_numeric.corr()

plt.figure(figsize=(10,6))
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm")
plt.title("Matriz de correlaciÃ³n")
plt.show()
```

```python
# 9. Boxplots (detecciÃ³n de outliers)
# ================================

num_cols = df.select_dtypes(include=np.number).columns

for col in num_cols:
    plt.figure(figsize=(8,4))
    sns.boxplot(x=df[col])
    plt.title(f"Outliers en: {col}")
    plt.show()
```
