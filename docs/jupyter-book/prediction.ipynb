```python
import joblib
import pandas as pd
import numpy as np
import os
```

##  1. Rutas robustas

Se construyen rutas absolutas basadas en el directorio del proyecto para asegurar portabilidad y compatibilidad con scripts y notebooks.


```python
# Detectar ra铆z del proyecto
current_dir = os.getcwd()
repo_root = os.path.dirname(current_dir)

model_dir = os.path.join(repo_root, "models")
data_dir = os.path.join(repo_root, "data", "raw")

model_path = os.path.join(model_dir, "xgb_model.pkl")
preprocessor_path = os.path.join(model_dir, "preprocessor.pkl")

model_path, preprocessor_path
```

## 锔 2. Cargar modelo y preprocesador


```python
model = joblib.load(model_path)
preprocessor = joblib.load(preprocessor_path)
```

##  3. Funci贸n de predicci贸n

La funci贸n acepta:

- Un diccionario con los atributos de un veh铆culo  
- Un DataFrame con m煤ltiples registros  

Devuelve:

- Un array con las predicciones del modelo


```python
def make_prediction(input_data):
    """
    Acepta:
        - Un diccionario con datos de un coche
        - Un DataFrame con varios registros
    Devuelve:
        - Predicciones del modelo como array
    """

    if isinstance(input_data, dict):
        df = pd.DataFrame([input_data])
    elif isinstance(input_data, pd.DataFrame):
        df = input_data.copy()
    else:
        raise ValueError("input_data debe ser un dict o un DataFrame")

    processed = preprocessor.transform(df)
    prediction = model.predict(processed)

    return prediction
```

## И 4. Ejemplo de uso

Puedes modificar los valores para generar predicciones personalizadas.


```python
ejemplo = {
    "Year": 2018,
    "Mileage": 35000,
    "Brand": "Toyota",
    "Model": "Corolla",
    "Fuel_Type": "Gasoline",
    "Transmission": "Automatic",
    "Engine_Size": 1.8,
}

make_prediction(ejemplo)
```
